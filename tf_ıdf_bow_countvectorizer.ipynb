#BoW

```python
import pandas as pd
import numpy as np
import collections
import re
doc1 = 'Game of Thrones is an amazing tv series!'
doc2 = 'Game of Thrones is the best tv series!'
doc3 = 'Game of Thrones is so great'
```

```python
l_doc1 = re.sub(r"[^a-zA-Z0-9]", " ", doc1.lower()).split()
l_doc2 = re.sub(r"[^a-zA-Z0-9]", " ", doc2.lower()).split()
l_doc3 = re.sub(r"[^a-zA-Z0-9]", " ", doc3.lower()).split()
```

```python
wordset12 = np.union1d(l_doc1,l_doc2)
wordset =  np.union1d(wordset12,l_doc3)
print(wordset)
```

```python
def calculateBOW(wordset,l_doc):
  tf_diz = dict.fromkeys(wordset,0)
  for word in l_doc:
      tf_diz[word]=l_doc.count(word)
  return tf_diz
```

```python
bow1 = calculateBOW(wordset,l_doc1)
bow2 = calculateBOW(wordset,l_doc2)
bow3 = calculateBOW(wordset,l_doc3)
df_bow = pd.DataFrame([bow1,bow2,bow3])
df_bow.head()
```

# scikit-learn CountVectorizer()

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Assuming doc1, doc2, and doc3 are your documents
doc1 = 'Game of Thrones is an amazing tv series!'
doc2 = 'Game of Thrones is the best tv series!'
doc3 = 'Game of Thrones is so great'

vectorizer = CountVectorizer(stop_words='english', ngram_range=(2,2))  #stop wordleri deanlık çıkarabilriz
X = vectorizer.fit_transform([doc1, doc2, doc3])

df_bow_sklearn = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df_bow_sklearn.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Assuming doc1, doc2, and doc3 are your documents
doc1 = "This is the first document"
doc2 = "This document is the second document"
doc3 = "And this is the third one"

vectorizer = CountVectorizer(stop_words='english', ngram_range=(2,2))    #kelime çiftleri kullanarak
X = vectorizer.fit_transform([doc1, doc2, doc3])

df_bow_sklearn = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df_bow_sklearn.head()
```

**Bow vs  Count Vectorizer()**

**CountVectorizer:** Bu, metin verilerini temsil etmek için sıklıkla kullanılan bir vektörleştirme aracıdır. Temelde belirli bir metin dökümanındaki kelime sayısını sayar ve bu sayıları bir vektörle temsil eder. Dolayısıyla, her bir özellik (sütun) bir kelimenin dökümandaki sayısını gösterir. CountVectorizer, kelime sayılarını dikkate alır, ancak bu kelimelerin belirli bir belgedeki göreceli önemini yansıtmaz.

**Bag of Words (BoW):** Bu kavram, bir metni temsil etmenin bir yolunu ifade eder. BoW modeli, bir metni bir koleksiyon (veya çanta) olarak ele alır, kelime sırasını göz ardı eder ve her bir kelimenin sıklığını belirler. Bu sıklıklar, genellikle bir vektörde temsil edilir ve her bir özellik bir kelimeye karşılık gelir. Yani, BoW, metni kelime seviyesinde temsil eden bir modeldir.

**Özetle**, CountVectorizer, BoW modelinin bir uygulamasıdır. CountVectorizer, BoW modelini uygulamak için bir araçtır ve bir metin belgesini bir vektörle temsil etmek için kullanılır, bu vektörde her bir özellik belirli bir kelimenin dökümandaki sayısını gösterir.

**Dil Modeli vs Kelime  Temsili**

**Dil Modeli:** Bir dil modeli, bir metindeki dil yapısını veya dilin olasılıklı yapılarını modellemeye çalışır. Bu modeller, bir kelimenin bir metin içindeki olasılığını tahmin etmek için kullanılabilir. Dil modelleri, genellikle bir metni verildiğinde bir sonraki kelimeyi tahmin etmek veya bir cümlenin akıcılığını değerlendirmek için kullanılır. Örneğin, bir metnin sonraki kelimesini tahmin eden bir dil modeli, bir dizi kelimenin ardışık olasılıklarını modelleyebilir.

**Kelime Temsili:** Kelime temsili, bir kelimenin bir vektörle temsil edilmesidir. Bu temsil, kelimenin semantik veya sözdizimsel özelliklerini yakalamaya çalışır. Örnek olarak, Word2Vec, GloVe ve FastText gibi yöntemler, kelimeleri vektör uzayında temsil etmek için kullanılan popüler tekniklerdir. Bu temsil, kelimenin anlamını daha iyi anlamak veya kelime benzerliğini ölçmek için kullanılabilir.

Dolayısıyla, bir dil modeli, bir metnin dil yapısını modellemek için kullanılırken, kelime temsili, kelimelerin vektör uzayında temsil edilmesi için kullanılan bir kavramdır. Bir dil modeli, metindeki kelimelerin dizisini ele alırken, kelime temsili, kelime düzeyinde temsil edilir ve daha genel olarak kullanılabilir.
